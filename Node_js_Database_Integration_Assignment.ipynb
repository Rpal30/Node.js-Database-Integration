Question 1: Using the `mysql2` library, write a Node.js script to establish a connection to a local
MySQL database named `app_db`.


```python
// Import the mysql2 library
const mysql = require('mysql2');

// Create a connection to the database
const connection = mysql.createConnection({
  host: 'localhost',     // Database host
  user: 'root',          // Your MySQL username
  password: 'password',  // Your MySQL password
  database: 'app_db'     // Database name
});

// Establish the connection
connection.connect((err) => {
  if (err) {
    console.error('Error connecting to the database:', err.message);
    return;
  }
  console.log('Connected to the MySQL database: app_db');
});

// Close the connection when done
// connection.end();
```

Key Notes :

Replace "root" and "password" with your actual MySQL credentials.

The connection.end() line is commented out so the connection stays open for queries. Uncomment it when you want to close the connection.

This script assumes MySQL is running locally on the default port (3306).

Question 2: Using an active mysql2 connection, write a function to fetch all rows from a
products table and log the results.

```python
// Function to fetch all products
function fetchAllProducts(connection) {
  const query = 'SELECT * FROM products';

  connection.query(query, (err, results) => {
    if (err) {
      console.error('Error fetching products:', err.message);
      return;
    }
    console.log('Products:', results);
  });
}

// Example usage:
// Assuming `connection` is already established with mysql2
// fetchAllProducts(connection);
```

How it works :

The function takes an existing connection object (created with mysql2.createConnection).

It runs a simple SELECT * FROM products query.

Results are returned as an array of row objects, which are logged to the console.

Errors are caught and logged for easier debugging.

Question 3: Define a Sequelize model for a Task that has a title (string) and a completed
(boolean) field.

```python
// Import Sequelize
const { Sequelize, DataTypes } = require('sequelize');

// Initialize Sequelize (example: local SQLite or MySQL)
const sequelize = new Sequelize('app_db', 'root', 'password', {
  host: 'localhost',
  dialect: 'mysql' // or 'sqlite', 'postgres', 'mssql'
});

// Define the Task model
const Task = sequelize.define('Task', {
  title: {
    type: DataTypes.STRING,
    allowNull: false
  },
  completed: {
    type: DataTypes.BOOLEAN,
    defaultValue: false
  }
}, {
  tableName: 'tasks',   // optional: specify table name
  timestamps: true      // adds createdAt and updatedAt fields
});

// Sync the model with the database
sequelize.sync()
  .then(() => {
    console.log('Task model synced with database');
  })
  .catch(err => {
    console.error('Error syncing model:', err);
  });

module.exports = Task;
```

Key Points :

DataTypes.STRING defines the title field as a string.

DataTypes.BOOLEAN defines the completed field as a boolean, with a default of false.

sequelize.sync() ensures the table is created if it doesn’t already exist.



Question 4: Using a Sequelize model named Task, write a query to find all tasks and log their
titles.


```python
// Assuming Task model is already defined and synced

async function logAllTaskTitles() {
  try {
    const tasks = await Task.findAll({
      attributes: ['title'] // fetch only the title field
    });

    tasks.forEach(task => {
      console.log(task.title);
    });
  } catch (err) {
    console.error('Error fetching tasks:', err.message);
  }
}

// Example usage
logAllTaskTitles();
```

Explanation :

Task.findAll() retrieves all rows from the tasks table.

The attributes option ensures only the title column is fetched, making the query more efficient.

The results are iterated with forEach to log each title individually.

Wrapping in try/catch ensures errors are handled gracefully.

Question 6: Write the code to connect to a local MongoDB server and a database named
inventory using the official mongodb driver.

```python
// Import the MongoDB driver
const { MongoClient } = require('mongodb');

// Connection URL for local MongoDB
const url = 'mongodb://localhost:27017';

// Database name
const dbName = 'inventory';

// Create a new MongoClient
const client = new MongoClient(url);

async function main() {
  try {
    // Connect to the MongoDB server
    await client.connect();
    console.log('Connected successfully to MongoDB server');

    // Select the database
    const db = client.db(dbName);

    // Example: list collections
    const collections = await db.listCollections().toArray();
    console.log('Collections:', collections);
  } catch (err) {
    console.error('Error connecting to MongoDB:', err.message);
  } finally {
    // Close the connection
    await client.close();
  }
}

// Run the main function
main();
```

Key Points :

MongoClient is the entry point for connecting to MongoDB.

The connection string mongodb://localhost:27017 assumes MongoDB is running locally on the default port.

client.db('inventory') selects the database named inventory.

Always wrap connections in try/catch and close them with client.close() to avoid resource leaks.

Question 7: Using the native MongoDB driver, insert a new document into a users collection
with a name of "John Doe" and an age of 30.


```python
// Import the MongoDB driver
const { MongoClient } = require('mongodb');

// Connection URL
const url = 'mongodb://localhost:27017';

// Database name
const dbName = 'inventory';

// Create a new MongoClient
const client = new MongoClient(url);

async function run() {
  try {
    // Connect to the MongoDB server
    await client.connect();
    console.log('Connected to MongoDB');

    // Select the database
    const db = client.db(dbName);

    // Select the users collection
    const usersCollection = db.collection('users');

    // Insert a new document
    const result = await usersCollection.insertOne({
      name: 'John Doe',
      age: 30
    });

    console.log('Inserted document with _id:', result.insertedId);
  } catch (err) {
    console.error('Error inserting document:', err.message);
  } finally {
    // Close the connection
    await client.close();
  }
}

// Run the function
run();
```

Key Points :

db.collection('users') targets the users collection.

insertOne() adds a single document with the specified fields.

The result.insertedId gives you the unique _id MongoDB assigned to the new document.



Question 8: Write a MongoDB query to find all users in a collection who are older than 25.


```python
// Import MongoDB driver
const { MongoClient } = require('mongodb');

// Connection URL
const url = 'mongodb://localhost:27017';

// Database name
const dbName = 'inventory';

const client = new MongoClient(url);

async function run() {
  try {
    await client.connect();
    console.log('Connected to MongoDB');

    const db = client.db(dbName);
    const usersCollection = db.collection('users');

    // Query: find all users older than 25
    const cursor = usersCollection.find({ age: { $gt: 25 } });

    const results = await cursor.toArray();
    console.log('Users older than 25:', results);
  } catch (err) {
    console.error('Error querying users:', err.message);
  } finally {
    await client.close();
  }
}

run();
```

Key Points :

The query uses { age: { $gt: 25 } }, where $gt means “greater than.”

find() returns a cursor, which you can iterate or convert to an array with toArray().



Question 9: Write a MongoDB query to update a user's age to 31, finding the user by their
name, "John Doe".

```python
// Import MongoDB driver
const { MongoClient } = require('mongodb');

// Connection URL
const url = 'mongodb://localhost:27017';

// Database name
const dbName = 'inventory';

const client = new MongoClient(url);

async function run() {
  try {
    await client.connect();
    console.log('Connected to MongoDB');

    const db = client.db(dbName);
    const usersCollection = db.collection('users');

    // Update query: find by name and set age to 31
    const result = await usersCollection.updateOne(
      { name: 'John Doe' },          // filter
      { $set: { age: 31 } }          // update
    );

    console.log(`Matched ${result.matchedCount} document(s) and updated ${result.modifiedCount} document(s).`);
  } catch (err) {
    console.error('Error updating user:', err.message);
  } finally {
    await client.close();
  }
}

run();
```

Key Points :

Filter: { name: 'John Doe' } ensures only the user with that name is targeted.

Update: { $set: { age: 31 } } modifies the age field to 31.

Result object: matchedCount tells how many documents matched the filter, and modifiedCount tells how many were actually updated.

Question 10: Write a MongoDB query to delete the user with the name "John Doe".


```python
// Import MongoDB driver
const { MongoClient } = require('mongodb');

// Connection URL
const url = 'mongodb://localhost:27017';

// Database name
const dbName = 'inventory';

const client = new MongoClient(url);

async function run() {
  try {
    await client.connect();
    console.log('Connected to MongoDB');

    const db = client.db(dbName);
    const usersCollection = db.collection('users');

    // Delete query: remove user with name "John Doe"
    const result = await usersCollection.deleteOne({ name: 'John Doe' });

    console.log(`Deleted ${result.deletedCount} document(s).`);
  } catch (err) {
    console.error('Error deleting user:', err.message);
  } finally {
    await client.close();
  }
}

run();
```

Key Points :

Filter: { name: 'John Doe' } ensures only the user with that exact name is targeted.

deleteOne() removes the first matching document. If multiple users share the same name, only one will be deleted.

The result.deletedCount tells you how many documents were actually removed.
